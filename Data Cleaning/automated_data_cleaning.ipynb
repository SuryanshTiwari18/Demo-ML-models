# Automated Data Cleaning

#### Types of Missing Data:

#### 1. Missing Completely at Random (MCAR)
#### 2. Missing At Random (MAR)
#### 3. Missing Not At Random (MNAR)

```python
import pandas as pd
```

```python
df=pd.read_csv('unclean_smartwatch_health_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['User ID'].isnull().sum()
```

```python
df=df.drop(columns=['User ID'])
```

```python
df.info()
```

```python
df['Heart Rate (BPM)'].isnull().sum()
```

```python
df[df['Heart Rate (BPM)'].isnull()==True]
```

```python
## No pattern with any column for Heart Rate (BPM) null values, so it comes under 
# the category of Missing Completely at Random (MCAR)
```

```python
### drop missing data --> dropna() ---> use only when the missing data is less 
# than 5% of the total data, otherwise it will lead to loss of information

### since overall null data is somwhere around 14% of the total data, we will not
#  drop the null values and instead find another way
```

```python
### Mean/Median Imputation --> fill the missing data with the mean or median of 
# all rows of that data

### while going with mean value ---> check for outliers in the data, if there are
#  outliers then go with median value instead of mean value
```

```python
### Age--->500(Outlier)---->1000--->mean---->??
### Age--->Median---->if you will arrange all values from low to high side
### Age--->Mode--->most frequently occuring value
```

```python
df['Heart Rate (BPM)'].describe()
```

```python
df['Heart Rate (BPM)']=df['Heart Rate (BPM)'].fillna(df['Heart Rate (BPM)'].median())
```

```python
df['Heart Rate (BPM)'].isnull().sum()
```

```python
df.info()
```

```python
df['Blood Oxygen Level (%)'].isnull().sum()
```

```python
df[df['Blood Oxygen Level (%)'].isnull()==True]
```

```python
df.describe()
```

```python
df['Blood Oxygen Level (%)']=df['Blood Oxygen Level (%)'].fillna(df['Blood Oxygen Level (%)'].mean())
```

```python
df.info()
```

```python
### Forward filling or backward filling  ----> time series data
```

```python
df.info()
```

```python
df[df['Step Count'].isnull()==True]
```

```python
### method ---> ffill() or bfill() ---> forward fill or backward fill

### error occured with 'method' so use .ffill() or .bfill() with the column name instead of the whole dataframe

df['Step Count'] = df['Step Count'].ffill()
```

```python
df['Step Count'].isnull().sum()
```

```python
df.info()
```

```python
df['Sleep Duration (hours)'].value_counts()
```

```python
df[df['Sleep Duration (hours)']=='ERROR']
```

```python
### object type data ---> convert to numeric type data
### to_numeric()
```

```python
df['Sleep Duration (hours)']=pd.to_numeric(df['Sleep Duration (hours)'], errors='coerce')
```

```python
df[df['Sleep Duration (hours)']=='ERROR']
```

```python
df['Sleep Duration (hours)'].dtype
```

```python
df['Sleep Duration (hours)']
```

```python
df['Sleep Duration (hours)'].mean()
```

```python
df['Sleep Duration (hours)'].median()
```

```python
df['Sleep Duration (hours)']=df['Sleep Duration (hours)'].fillna(df['Sleep Duration (hours)'].mean())
```

```python
df['Sleep Duration (hours)'].isnull().sum()
```

```python
df.head()
```

```python
### For categorical columns ---> Mode ---> Most frequently occuring value in your column
```

```python
df['Activity Level'].mode()
```

```python
df['Activity Level'].isnull().sum()
```

```python
df['Activity Level']=df['Activity Level'].fillna('Seddentary')
```

```python
df.info()
```

```python
### constant imputation ---> when you want to fill the missing values with the constant values like 'unknown'
```

```python
df['Stress Level'].value_counts()
```

```python
### since very high value is required here we will go with constant imputation
```

```python
df['Stress Level']=df['Stress Level'].fillna(5)
```

```python
df.info()
```

```python
df=pd.read_csv('unclean_smartwatch_health_data.csv')
```

```python
df.info()
```

```python
df=df.drop(columns=['User ID'])
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
```

```python
df['Heart Rate (BPM)']=imputer.fit_transform(df[['Heart Rate (BPM)']])
```

```python
df.info()
```

```python
df.describe()
```

```python
imputer=SimpleImputer(strategy='mean')
```

```python
df['Blood Oxygen Level (%)']=imputer.fit_transform(df[['Blood Oxygen Level (%)']])
```

```python
imputer=SimpleImputer(strategy='most_frequent')
```

```python
df['Activity Level']=imputer.fit_transform(df[['Activity Level']]).ravel()
```

```python
df.info()
```

```python
imputer=SimpleImputer(strategy='constant', fill_value=5)
```

```python
df['Stress Level']=imputer.fit_transform(df[['Stress Level']]).ravel()
```
