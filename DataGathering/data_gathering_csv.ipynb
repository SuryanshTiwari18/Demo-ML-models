### 1. How to Read content from a csv file

```python
import pandas as pd
```

```python
pd.read_csv('Titanic-Dataset.csv')
```

#### 2. names ---> name is going to take a list of names that we can provide to our column names 
#### ---> names=['Name','Age']

```python
pd.read_csv('Titanic-Dataset.csv',names=['P_ID', 'Survived','Pclass','Name','Gender','Age','Sibling or Spouse','Parch','Ticket_ID','Fare','Cabin','Starting Journey'])
```

### header ---> header parameter makes sure that any of the row name will be shown as the column name only based on indexing of rows

```python
pd.read_csv('Titanic-Dataset.csv')
```

### index_col ---> keep one of our column as the index

```python
pd.read_csv('Titanic-Dataset.csv',index_col='PassengerId')
```

### usecols ---> what columns we should use to show out of all of the columns we are having in the dataframe
###  ---> use a list to mention all of the column names which are required

```python
pd.read_csv('student_depression_dataset.csv',usecols=['id','Gender','Age','City','Profession','Academic Pressure','CGPA','Study Satisfaction','Sleep Duration','Dietary Habits','Degree','Have you ever had suicidal thoughts ?','Depression'])
```

### squeeze ---> DataFrame ---> Series ---> squeeze ---> True ---> combination of usecols and squeeze ---> True

```python
pd.read_csv('Titanic-Dataset.csv',usecols=['Survived']).squeeze()
```

```python
pd.read_csv('student_depression_dataset.csv')
```

### nrows ---> it is going to help us with a fixed number of rows out of the complete data ---> 100 ---> only 100 rows out of all rows will be shown 

```python
pd.read_csv('student_depression_dataset.csv',nrows=1000)
```

### dtypes --->it can assign initially only with a specific data type to a column --->
### dtype= {'col_1':int, 'col_2':float}

```python
pd.read_csv('student_depression_dataset.csv',dtype={'Age':int})
```

### parse_dates = ['col_1'] ---> to convert the data type of date column to datetime

```python
pd.read_csv('Titanic-Dataset.csv')
```

### converter

```python
def station(value):
    if value == 'Q':
        return 'Queenstown'
    else:
        return value
```

```python
pd.read_csv('Titanic-Dataset.csv',converters={'Embarked':station})
```

### na_values

### it can help us out with mentioning what value to be conidered as a missing value

### na_values=['',':','No']

```python
pd.read_csv('Titanic-Dataset.csv',na_values=['NA','Missing','Not Available'])
```
