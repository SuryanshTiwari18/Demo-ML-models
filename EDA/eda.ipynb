### steps for EDA

#### Categories the given column as a Numerical, Categorical, Mixed Data type column
#### Univariate Analysis ----> Numerical or Categorical column
#### Numerical column analysis ---> Distribution of the data ---> histogram, kdeplot
#### ----> Outliers ---> Boxplot
#### ----> Handle missing values

#### categorical column analysis ----> distribution of the data ----> countplot
#### ----> missing values ---->


#### Bivariate Analysis ----> Numerical Numerical ---> scatterplot
####   ----> Numerical Categorical ----> barplot
####   ----> Categorical Categorical ----> Heatmap

#### EDA ----> Feature Engineering 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('tested.csv')
```

```python
df.head()
```

```python
## Univariate Analysis on Numerical Column --->
## Data Distribution of the numerical column --->describe() & plotting ---> kdeplot or histogram
## Outlier Detection ---> boxplot
```

```python
df['Age'].describe()
```

```python
df[df['Age']==76]
```

```python
sns.boxplot(data=df,x='Age')
```

```python
## Since the two points are true therefore no outlier treatment is required.
```

```python
df['Age'].isnull().sum()
```

```python
## Here we will use median imputation to fill the missing values in the Age column. 
# Since the distribution of Age is skewed, median is a better measure of central tendency than mean.
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df['Age'].isnull().sum()
```

```python
df['Age'].describe()
```

```python
sns.histplot(data=df,x='Age')
```

```python
sns.kdeplot(df['Age'])
```
