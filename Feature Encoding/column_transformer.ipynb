# Column Transformer

```python
import pandas as pd
```

```python
df=pd.read_csv('Bank_Marketing_Original_Data.csv')
```

```python
df.head()
```

```python
## Transformers ---> OrdinalEncoder, OneHotEncoder, SimpleImputer
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(
    transformers=[
        ('tr1', OrdinalEncoder(categories=[['unknown', 'primary', 'secondary', 'tertiary']]), ['Education']),
        ('tr2', OneHotEncoder(), ['Job', 'Marital', 'Default', 'Housing', 'Loan', 'contact', 'Month', 'poutcome'])
    ], 
    remainder='passthrough'
)
```

```python
transformer
```

```python
X=df.iloc[:, :-1]
```

```python
Y=df['TARGET']
```

```python
X=transformer.fit_transform(X)
```
