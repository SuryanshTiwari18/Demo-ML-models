```python
## Ordinal Column ----> Education ---> Ordinal encoder ---> input data must be in 2D format
## Label Encoder ---> Target Column
## Nominal Column ----> rest of the categorical columns
```

```python
ordinal_data=bank[['Education']]
```

```python
target_data=bank[['TARGET']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[['unknown','primary','secondary','tertiary']])
```

```python
ordinal_data=oe.fit_transform(ordinal_data)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
target_data=le.fit_transform(target_data)
```
