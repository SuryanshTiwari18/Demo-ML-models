```python
text="<h1>Hello World</h1>"
```

```python
import re
```

```python
re.sub(r'<.*?>', '', text)
## r means it is a raw string, <.*?> means any character between < and >, ? means non-greedy match, '' means replace with empty string
```

```python
data=pd.DataFrame({"col":["<h1>Hello World</h1>","<a>Google</a>","<p>This is a paragraph.</p>"]})
```

```python
data
```

```python
def html_cleaning(text):
    empty_html = []
    for i in text:
        clean_text = re.sub(r'<.*?>', '', i)
        empty_html.append(clean_text)
    return empty_html
```

```python
data.apply(html_cleaning)
```

```python
slang_dict = {"asap": "as soon as possible", "gn": "good night", "lol": "laugh out loud", "yolo": "you only live once"}
```

```python
slang=pd.DataFrame({"slang":["Hey call me asap","hey hello gn","lol that was funny","hey yolo, you know it"]})
```

```python
text="Hey call me asap"
```

```python
def expand_slang(text,slang_dict):
    words=text.split()
    return " ".join([slang_dict.get(word,word) for word in words])
```

```python
expand_slang(text,slang_dict)
```
