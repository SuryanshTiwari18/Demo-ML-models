```python
text="<h1>Hello World</h1>"
```

```python
import re
```

```python
re.sub(r'<.*?>', '', text)
## r means it is a raw string, <.*?> means any character between < and >, ? means non-greedy match, '' means replace with empty string
```

```python
data=pd.DataFrame({"col":["<h1>Hello World</h1>","<a>Google</a>","<p>This is a paragraph.</p>"]})
```

```python
data
```

```python
def html_cleaning(text):
    empty_html = []
    for i in text:
        clean_text = re.sub(r'<.*?>', '', i)
        empty_html.append(clean_text)
    return empty_html
```

```python
data.apply(html_cleaning)
```

```python
def expand_slang(text,slang_dict):
    words=text.split()
    return " ".join([slang_dict.get(word,word) for word in words])
```

```python
## Removing emoji --> spam detection / topic modelling
```

```python
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
data={
    'content':[
        "I love pizza 🍕 so much!",
        "Good morning! ☀️",
        "Let's go for a ride 🚗",
        "Vacation time 🏖️",
        "Happy birthday! 🎉",
        "Feeling strong 💪"
    ]
}

data = pd.DataFrame(data)
```

```python
data
```

```python
data['content'].apply(remove_emoji)
```
