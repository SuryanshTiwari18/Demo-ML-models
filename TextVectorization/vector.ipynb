```python
d=pd.DataFrame({'col': ['People Love AI', 'AI Love data', 'data Love Love', 'Love People People']})
```

```python
d
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
d['col']
```

```python
cv.fit_transform(d['col']).toarray()
```

```python
cv.vocabulary_
```

```python
cv=CountVectorizer(ngram_range=(2,2))
```

```python
cv.fit_transform(d['col']).toarray()
```

```python
cv.vocabulary_
```

```python
len(cv.vocabulary_)
```
