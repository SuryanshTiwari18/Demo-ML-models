```python
## Applying the NLP Preprocessing ---> 1. Tokenization ---> Sentence Tokenization 2. Stopword removal ---> we don't want embedding of unnecessary words 3. Lowercasing and the rest of the basic preprocessing steps
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
potter=[]
data=sent_tokenize(data)
```

```python
len(data)
```

```python
for i in data:
    potter.append(simple_preprocess(i))
```

```python
from nltk.corpus import stopwords
```

```python
stopwords=set(stopwords.words('english'))
```

```python
stopwords
```

```python
def remove_stopwords(text):
    cleaned_text=[]
    for sentence in text:
        cleaned_sentence=[]
        for word in sentence:
            if word.lower() not in stopwords:
                cleaned_sentence.append(word)
        cleaned_text.append(cleaned_sentence)
    return cleaned_text
```

```python
cleaned_potter=remove_stopwords(potter)
```

```python
import gensim
```

```python
model=gensim.models.Word2Vec(window=10,min_count=2)
```

```python
model.build_vocab(cleaned_potter)
```

```python
model.train(cleaned_potter, total_examples=model.corpus_count, epochs=model.epochs)
```
